```python
#np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0, 75,50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([110, 180, 240, 99, 220])
```

```python
font1 = {'family':'serif','color':'indigo','size':15}
font2 = {'family':'monospace','color':'green','size':14}


fig,(ax1, ax2) = plt.subplots(ncols=2, 
                              nrows=1, 
                              figsize=(7, 3))

ax1.scatter(x1, y1, c='purple', s=50, alpha=0.6, edgecolors='black')
ax1.set_title("Scatter Plot", fontdict=font1)
ax1.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.xlabel('X Values', fontdict=font1)
ax2.bar(x2, y2, color='indigo', edgecolor='black')
ax2.set_title("Bar Plot", fontdict=font2)
ax2.xaxis.set_tick_params(rotation=45, labelsize=12)
ax2.legend(["Characters"], loc='upper left', fontsize=10, bbox_to_anchor=(1,1))
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],     #define the layout of the subplots in a mosaic pattern
                                      ['ax2', 'ax3']],  #specify the arrangement of the subplots in a 2x2 grid, where 'ax1' and 'ax2' occupy the first row and 'ax3' occupies the second row
                                      figsize=(7, 4))

someaxes['ax1'].scatter(x,y1)       #create a scatter plot on the subplot labeled 'ax1' using the x and y1 data
someaxes['ax2'].bar(x2, y2)     #create a bar plot on the subplot labeled 'ax2' using the x2 and y2 data
someaxes['ax3'].plot(x,y1)      #create a line plot on the subplot labeled 'ax3' using the x and y1 data

someaxes['ax1'].set_xlabel('A Big label', fontsize=18) #create a messy look by having large font sizes for all three charts. Note how first label isn't displayed altogether
someaxes['ax2'].set_xlabel('Another Label', fontsize=18)
someaxes['ax3'].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize=18)
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],     #define the layout of the subplots in a mosaic pattern
                                      ['ax2', 'ax3']],  #specify the arrangement of the subplots in a 2x2 grid, where 'ax1' and 'ax2' occupy the first row and 'ax3' occupies the second row
                                      figsize=(7, 4),
                                      layout='constrained') #add layout='constrained' to automatically adjust the spacing between subplots to prevent overlap of elements

someaxes['ax1'].scatter(x,y1)       #create a scatter plot on the subplot labeled 'ax1' using the x and y1 data
someaxes['ax2'].bar(x2, y2)     #create a bar plot on the subplot labeled 'ax2' using the x2 and y2 data
someaxes['ax3'].plot(x,y1)      #create a line plot on the subplot labeled 'ax3' using the x and y1 data

someaxes['ax1'].set_xlabel('A Big label', fontsize=18) #create a messy look by having large font sizes for all three charts. Note how with layout='constrained', the first label is now displayed and there is no overlap of elements
someaxes['ax2'].set_xlabel('Another Label', fontsize=18)
someaxes['ax3'].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize=18)
```

```python
# first make our sample data
x = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])

# define our figure and axes (just one this time)
fig, ax = plt.subplots(figsize=(7, 3))

# now call both bar and plot elements to the same axes (ax)
ax.bar(x, y1,color = "indigo")
ax.plot(x, y2,color = "red")
```

```python
y2_sd = np.std(y2)

fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")

ax.errorbar(x,
            y2,
            yerr=y2_sd,
            fmt="none",
            ecolor="indigo",
            elinewidth=2,
            capsize=6,
            capthick=4
            )
```

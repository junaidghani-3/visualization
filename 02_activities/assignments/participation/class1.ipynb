```python
# importing the libraries
import numpy as np      # for numerical computations
import matplotlib.pyplot as plt     # for data visualization
import pandas as pd         # for data manipulation and analysis
```

```python
# creating the data

np.random.seed(613)                         # for reproducibility       
x = np.arange(50)                           # x values from 0 to 49
y = np.random.randint(0, 100, 50)           # 50 random integers between 0 and 100

# plotting the data
fig, ax = plt.subplots(figsize=(5, 3))      # Set the figure size
#ax.scatter(x, y)                           # scatter plot
#ax.bar(x, y)                               # bar plot
#ax.barh(x, y)                              # horizontal bar plot
#ax.plot(x, y)                              # line plot
ax.hist(y)                                  # histogram
```

```python
# customizing fonts
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}

# plot with customized fonts
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y)
ax.set_title('Total growth over time', fontdict=font1, loc='left')
ax.set_ylabel('Total growth', fontdict=font2)
ax.set_xlabel('Years since start', fontdict=font2)
fig.tight_layout()
```

```python
# scatter plot with customized marker and color

fig, ax = plt.subplots(figsize=(5, 3)) 
ax.scatter(x, y,marker = '*', color = 'indigo')
```

```python
# line plot with customized marker, color, line style, and line width

fig, ax = plt.subplots(figsize=(5, 3)) 
ax.plot(x, y,marker='*', color = 'indigo', linestyle='--', linewidth=2)
```

```python
# line plot with customized marker, color, line style, line width, and marker edge and face color

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y,
        marker='*', 
        markersize = 12, 
        color = '#7425b9', 
        linestyle = '--', 
        linewidth = 2, 
        markeredgecolor = '#fa9359', 
        markerfacecolor = '#000000' )
ax.grid()
```

```python
# line plot with customized marker, color, line style, line width, and marker edge and face color

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y,
        marker='o', 
        markersize = 12, 
        color = '#7425b9', 
        linestyle = '--', 
        linewidth = 2, 
        markeredgecolor = '#fa9359', 
        markerfacecolor = '#000000' )
ax.grid(axis='y', color = 'lightblue', linestyle = '--', linewidth = 2)
ax.grid(axis='x', color = 'red', linestyle = 'dotted', linewidth = 1)
```
